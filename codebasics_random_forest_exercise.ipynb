```python
import pandas
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

(iris_features, iris_targets) = load_iris (return_X_y = True, as_frame = True)
```

```python
iris_targets.head(15)

# iris_targets.head(51) # Uncomment to see that the flower varieties are sorted AND encoded.
```

```python
# We have 3 varieties of flowers, all in equal amounts.  There's 150 flowers in all
print (iris_targets.value_counts())
```

```python
# Horray, no missing row data!  We can jump straight to model construction for this exercise.
print (iris_features.isnull().sum())
```

```python
features_train, features_test, targets_train, targets_test = train_test_split (\
  iris_features, iris_targets, test_size = 0.2, random_state = 5)
```

```python
# Always scale your numerical measurements before training your model to them.
iris_scaler = StandardScaler(with_std = True)
iris_scaler.fit (features_train.values)
scaled_features_train = iris_scaler.transform (features_train.values)
scaled_features_test = iris_scaler.transform (features_test.values)
```

```python
iris_classifier_10 = RandomForestClassifier (random_state = 5)
iris_classifier_10.fit (features_train, targets_train)
print (iris_classifier_10.score (features_test, targets_test)) # 93% accuracy
```

```python
iris_classifier_2 = RandomForestClassifier (n_estimators = 2, random_state = 5)
iris_classifier_2.fit (features_train, targets_train)
print (iris_classifier_2.score (features_test, targets_test)) # 90% accuracy with only two trees
```

```python
iris_classifier_250 = RandomForestClassifier (n_estimators = 250, random_state = 5)
iris_classifier_250.fit (features_train, targets_train)
print (iris_classifier_250.score (features_test, targets_test)) # same accuracy with 250 trees as with 100...
# My theory is that accuracy doesn't change much beyond 100 trees because we're
# sampling data from a small dataset and there's only so many decision trees you can make.
```
